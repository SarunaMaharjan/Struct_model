```python
from transformers import XLMRobertaModel, AutoTokenizer

model_name = "xlm-roberta-base"
save_directory = "./xlm-roberta-base-local"

print(f"Downloading model and tokenizer for {model_name}...")
# This part remains the same
model = XLMRobertaModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

print(f"Saving model and tokenizer to {save_directory}...")
# --- THE FIX IS HERE ---
# We add `safe_serialization=False` to use the older saving method.
model.save_pretrained(save_directory, safe_serialization=False)
tokenizer.save_pretrained(save_directory)

print("Download complete!")
```
